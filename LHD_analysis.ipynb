# Growth rates 

```python
def reader():
    global filename

    data = [ (pylab.loadtxt("{0}".format(filename))) ][0]
    k = data[:, 1]
    g = data[:, 2]
    w = data[:, 3]

    return k, g
```

```python
%cd /raven/u/cmor/jupyter/HP_project/data/LHD/growth
filename = 'kyscanalpha0npol1.txt'
k_a_0_p_1, g_a_0_p_1 = reader()

filename = 'kyscanalpha0npol2.txt'
k_a_0_p_2, g_a_0_p_2 = reader()

filename = 'kyscanalpha0npol3.txt'
k_a_0_p_3, g_a_0_p_3 = reader()

filename = 'kyscanalphapiover10npol1.txt'
k_a_10_p_1, g_a_10_p_1 = reader()

filename = 'kyscanalphapiover10npol2.txt'
k_a_10_p_2, g_a_10_p_2 = reader()

filename = 'kyscanalphapiover10npol3.txt'
k_a_10_p_3, g_a_10_p_3 = reader()
```

```python
plt.figure()
plt.plot(k_a_0_p_1, g_a_0_p_1, '-o', label=r"$\alpha:0$, turns:1", linewidth=1.5)
plt.plot(k_a_0_p_2, g_a_0_p_2, '-o', label=r"$\alpha:0$, turns:2", linewidth=1.5)
plt.plot(k_a_0_p_3, g_a_0_p_3, '-o', label=r"$\alpha:0$, turns:3", linewidth=1.5)
plt.plot(k_a_10_p_1, g_a_10_p_1, '-*', label=r"$\alpha:\pi / 10$, turns:1", linewidth=1)
plt.plot(k_a_10_p_2, g_a_10_p_2, '-*', label=r"$\alpha:\pi / 10$, turns:2", linewidth=1)
plt.plot(k_a_10_p_3, g_a_10_p_3, '-*', label=r"$\alpha:\pi / 10$, turns:3", linewidth=1)
plt.hlines(y=0.258, xmin=0, xmax=2.5, alpha=0.7, linewidth=1.5, color="red", label="Surface")
plt.title(r"Linear growth rate vs. wavenumber LHD")
plt.xlabel(r"$k_{y} \rho_{i}$")
plt.ylabel(r"\ $\gamma \ (c_s / a)$")
plt.legend(loc="right")
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.savefig("/u/cmor/jupyter/HP_project/figures/lhd_growth_rates.png")
plt.show()
```
